```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data1 = "out_HL=0.65_beta_0.01.dat"
df1 = pd.read_csv(data1, sep = "\t", skiprows = 6003, header = None)
```

```python
df1
```

```python
n = len(df1)
for i in range(n/6000):
    plt.plot(df1[i*6000:(i+1)*6000][0],df1[i*6000:(i+1)*6000][1])
#plt.axis([30,60, 0, 2])
plt.title("Beta = 0.01 b = 0.045692 D = 0.014613")
plt.xlabel("x")
plt.ylabel("Height")
plt.savefig('RUC1_beta_0.01.jpg')
plt.show()
```
