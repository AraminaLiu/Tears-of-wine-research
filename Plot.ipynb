```python
def find_maxrow(df):
    max = 0
    max_row = 0
    for i in range(len(df)):
        if (df[0][i] > 22) & (df[1][i] > max):
            max = df[1][i]
            max_row = i
            
    return max_row
```

```python
def find_minrow(df):
    min = 0.6
    min_row = 0
    for i in range(len(df)):
        if (df[0][i] > 22) & (df[1][i] < min):
            min = df[1][i]
            min_row = i
            
    return min_row
```

```python
max_time = []
max_height = []
max_xlabel = []
min_time = []
min_height = []
min_xlabel = []
for i in range(n/4000):
    df1 = df[i*4000:(i+1)*4000].reset_index(drop = True)
    maxrow = find_maxrow(df1)
    minrow = find_minrow(df1)
    max_time.append(df1[4][maxrow])
    max_height.append(df1[1][maxrow])
    max_xlabel.append(df[0][maxrow])
    min_time.append(df1[4][minrow])
    min_height.append(df1[1][minrow])
    min_xlabel.append(df1[0][minrow])

print max_time
print max_height
print max_xlabel
print min_time
print min_height
print min_xlabel
```

```python
def speed_calc(data):
    speed_list = []
    for i in range(len(data)-1):
        speed = (data.iloc[i+1, 2] - data.iloc[i, 2])/(data.iloc[i+1, 0] - data.iloc[i, 0])
        speed_list.append(speed)
    return speed_list
```

```python
speed_max = speed_calc(max_df)
speed_max
```

```python
speed_min =  speed_calc(min_df)
```
