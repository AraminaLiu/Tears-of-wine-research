```python
def find_maxrow(df):
    max = 0
    max_row = 0
    for i in range(len(df)):
        if (df[0][i] > 23) & (df[1][i] > max):
            max = df[1][i]
            max_row = i
            
    return max_row
```

```python
def find_minrow(df):
    min = 0.6
    min_row = 0
    for i in range(len(df)):
        if (df[0][i] > 23) & (df[1][i] < min):
            min = df[1][i]
            min_row = i
            
    return min_row
```

```python
def speed_calc(data):
    speed_list = []
    for i in range(len(data)-1):
        speed = (data.iloc[i+1, 2] - data.iloc[i, 2])/(data.iloc[i+1, 0] - data.iloc[i, 0])
        speed_list.append(speed)
    return speed_list
```

```python
def export_speed(df):
        max_time = []
        max_height = []
        max_xlabel = []
        min_time = []
        min_height = []
        min_xlabel = []
        n = len(df)
        for i in range(n/6000):
            df1 = df[i*6000:(i+1)*6000].reset_index(drop = True)
            maxrow = find_maxrow(df1)
            minrow = find_minrow(df1)
            max_time.append(df1[4][maxrow])
            max_height.append(df1[1][maxrow])
            max_xlabel.append(df1[0][maxrow])
            min_time.append(df1[4][minrow])
            min_height.append(df1[1][minrow])
            min_xlabel.append(df1[0][minrow])
        
        zippedList1 =  list(zip(max_time, max_height, max_xlabel))
        max_df = pd.DataFrame(zippedList1, columns = ('max_time', 'max_height', 'max_x'))
        
        zippedList2 =  list(zip(min_time, min_height, min_xlabel))
        min_df = pd.DataFrame(zippedList2, columns = ('min_time', 'min_height', 'min_x'))
        
        speed_max = speed_calc(max_df)
        speed_max = pd.DataFrame(speed_max, columns = ['speed_max'])
        
        speed_min = speed_calc(min_df)
        speed_min = pd.DataFrame(speed_min, columns = ['speed_min'])
        
        max_df1 = pd.concat([max_df[0:-1], speed_max], axis = 1)
        min_df1 = pd.concat([min_df[0:-1], speed_min], axis = 1)
        max_df1.to_csv('max_df.csv', index = False)
        min_df1.to_csv('min_df.csv', index = False)
```

```python
def show_speed():
    max_df = pd.read_csv('max_df.csv')
    min_df = pd.read_csv('min_df.csv')
        
    plt.plot(max_df.iloc[:, 0],max_df.iloc[:, 3])
    plt.plot(min_df.iloc[:, 0], min_df.iloc[:, 3])

    plt.title("Speed of curves", fontsize=16)
    plt.xlabel("Time", fontsize=14)
    plt.ylabel("Speed", fontsize=14)
    #plt.savefig('4_beta_0.03.jpg')
    plt.show()
```
